## Feature Engineering

```python
df.dtypes
```

```python
# Formatting enrolled_date column 
df["enrolled_date"] = pd.to_datetime(df.enrolled_date)
```

```python
df["difference"] = (df.enrolled_date-df.first_open).astype('timedelta64[h]')
```

```python
df.loc[df.difference > 48, "enrolled"].head()
```

```python
df.loc[df.difference > 48, "enrolled"] = 0
```

```python
df.loc[df.difference > 48, "enrolled"].head()
```

```python
df = df.drop(columns=["enrolled_date", "difference", "first_open"])
```

```python
df.head()
```

```python
## Formatting the screen_list Field
```

```python
# Load Top Screens
top_screens = pd.read_csv("top_screens.csv").top_screens.values
```

```python
top_screens
```

```python
## We'll add a comma to count the number of screens 
df["screen_list"] = df.screen_list.astype(str) + ","
```

```python
#We'll create a column for each top screen in the screen list 
## We drop top screens in screen list 

for sc in top_screens:
    df[sc] = df.screen_list.str.contains(sc).astype(int)
    df['screen_list'] = df.screen_list.str.replace(sc+",", "")
```

```python
##We'll create a new column with the count of the other screens
##We'll drop the screen_list column 
df["Other"] = df["screen_list"].str.count(",")
df = df.drop(columns=['screen_list'])
```

```python
pd.options.display.max_columns = 100
df.head()
```

```python
##We'll group the screens of the same category
##We'll drop the saving screens columns
savings_screens = ["Saving1", "Saving2", "Saving2Amount", "Saving4", "Saving5", 
                   "Saving6", "Saving7", "Saving8", "Saving9","Saving10"]
df["Saving_count"] = df[savings_screens].sum(axis=1)
df = df.drop(columns=savings_screens)
```

```python
##We'll group the screens of the same category
##We'll drop the Credit columns
credit_screens = ["Credit1", "Credit2", "Credit3", "Credit3Container", "Credit3Dashboard"]
df["CreditCount"] = df[credit_screens].sum(axis=1)
df = df.drop(columns=credit_screens)
```

```python
##We'll group the screens of the same category
##We'll drop the CC columns
cc_screens = ["CC1", "CC1Category", "CC3"]
df["CCCount"] = df[cc_screens].sum(axis=1)
df = df.drop(columns=cc_screens)
```

```python
##We'll group the screens of the same category
##We'll drop the Loan columns
loan_screens = ["Loan", "Loan2", "Loan3", "Loan4"]
df["LoansCount"] = df[loan_screens].sum(axis=1)
df = df.drop(columns=loan_screens)
```

## Data Pre-Processing 


```python
df = pd.read_csv('new_appdata.csv')
```

```python
plt.subplots(figsize=(10,6))
sns.countplot(df["enrolled"], label="count");
```

```python
# Splitting Independent and Response Variables
# Let's drop the target label coloumns
X = df.drop(["enrolled"], axis=1)
```

```python
X.head()
```

```python
y = df["enrolled"]
```

```python
y.head()
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
# Removing Identifiers
train_id = X_train["user"]
X_train = X_train.drop(columns=["user"])
test_id = X_test["user"]
X_test = X_test.drop(columns=["user"])
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

## Model Building

```python
# Fitting Model to the Training Set
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(penalty="l1", random_state=0)
```

```python
lr.fit(X_train_scaled, y_train)
```

```python
# Predicting Test Set
y_pred = lr.predict(X_test_scaled)
```

## Evaluating Results

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
```

```python
plt.subplots(figsize=(10,6))
sns.heatmap(cm, annot=True, fmt="d");
```

```python
print(classification_report(y_test, y_pred))
```

```python
print("Test Data Accuracy: %0.4f" % accuracy_score(y_test, y_pred))
```

## Applying k-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(lr, X_train_scaled, y_train, cv = 10)
print("Logistic Accuracy: %0.3f (+/- %0.3f)" % (accuracies.mean(), accuracies.std() * 2))
```

## Model Tuning

### Grid Search (Round 1)

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"penalty" : ["l1", "l2"], "C": [0.001, 0.01, 0.1, 1, 10, 100, 1000]} 
```

```python
grid = GridSearchCV(lr, param_grid,scoring = "accuracy", cv = 10, n_jobs = -1)
```

```python
t0 = time.time()
grid.fit(X_train_scaled,y_train)
t1 = time.time()
print("Took %0.2f seconds" % (t1 - t0))
```

```python
best_parameters = grid.best_params_  
print(best_parameters)  
```

```python
best_result = grid.best_score_  
print(best_result)  
```

## Grid Search (Round 2)


```python
param_grid = {"penalty" : ["l1", "l2"], "C": [0.1, 0.5, 0.9, 1, 2, 5]} 
```

```python
grid = GridSearchCV(lr, param_grid,scoring = "accuracy", cv = 10, n_jobs = -1)
```

```python
t0 = time.time()
grid.fit(X_train_scaled,y_train)
t1 = time.time()
print("Took %0.2f seconds" % (t1 - t0))
```

```python
best_parameters = grid.best_params_  
print(best_parameters)  
```

```python
best_result = grid.best_score_  
print(best_result)  
```
