## Data Preprocessing

```python
churn_data = pd.read_csv("clean_churn_data.csv")
```

```python
churn_data.head()
```

```python
user_id = churn_data["user"]
```

```python
user_id.head()
```

```python
churn_data = churn_data.drop(columns = ["user"])
```

```python
churn_data.head()
```

```python
# One-Hot Encoding
churn_data = pd.get_dummies(churn_data)
```

```python
churn_data.columns
```

```python
churn_data = churn_data.drop(columns = ["housing_na", "payment_type_na", "zodiac_sign_na"])
```

## Splitting the dataset into the Training set and Test set

```python
# Splitting Independent and Response Variables
# Let's drop the target label coloumns
X = churn_data.drop(["churn"], axis=1)
```

```python
X.head()
```

```python
y = churn_data["churn"]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

## Balancing the Training Set (Down-sample Majority Class)

```python
y_train.value_counts()
```

```python
pos_index = y_train[y_train.values == 1].index
neg_index = y_train[y_train.values == 0].index
```

```python
if len(pos_index) > len(neg_index):
    higher = pos_index
    lower = neg_index 
else:
    higher = neg_index
    lower = pos_index
```

```python
random.seed(0)
```

```python
higher = np.random.choice(higher, size=len(lower))
```

```python
lower = np.asarray(lower)
```

```python
new_indexes = np.concatenate((lower, higher))
```

```python
X_train = X_train.loc[new_indexes,]
y_train = y_train[new_indexes]
```

```python
X_train.shape, y_train.shape
```

```python
y_train.value_counts()
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train2 = pd.DataFrame(scaler.fit_transform(X_train))
X_test2 = pd.DataFrame(scaler.fit_transform(X_test))
```

```python
X_train2.columns = X_train.columns.values
X_test2.columns = X_test.columns.values
```

```python
X_train2.index = X_train.index.values 
X_test2.index = X_test.index.values
```

```python
X_train = X_train2
X_test = X_test2 
```

```python
X_train.head()
```

```python
X_test.head()
```

## Model Building

### Comparing Models

#### Logistic Regression

```python
# Fitting Model to the Training Set
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(random_state=0)
```

```python
lr.fit(X_train, y_train)
```

```python
# Predicting Test Set
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score, precision_score, recall_score
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```

```python
prec = precision_score(y_test, y_pred) # tp / (tp + fp)
prec
```

```python
rec = recall_score(y_test, y_pred) # tp / (tp + fn)
rec
```

```python
f1 = f1_score(y_test, y_pred)
f1
```

```python
df_cm = pd.DataFrame(cm, index = (0, 1), columns = (0, 1))
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)
sns.heatmap(df_cm, annot=True, fmt='g')
print("Test Data Accuracy: %0.4f" % acc)
```

```python
print(classification_report(y_test, y_pred))
```

```python
results = pd.DataFrame([["Logistic Regression", acc, prec, rec, f1]],
               columns = ["Model", "Accuracy", "Precision", "Recall", "F1 Score"])
```

```python
results
```

#### Support Vector Machine

```python
from sklearn.svm import SVC
svm = SVC(random_state = 0, kernel = 'rbf')
svm.fit(X_train, y_train)
```

```python
# Predicting Test Set
y_pred = svm.predict(X_test)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
df_cm = pd.DataFrame(cm, index = (0, 1), columns = (0, 1))
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)
sns.heatmap(df_cm, annot=True, fmt='g')
print("Test Data Accuracy: %0.4f" % acc)
```

```python
print(classification_report(y_test, y_pred))
```

```python
model_results = pd.DataFrame([['SVM (RBF)', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])

results = results.append(model_results, ignore_index = True)
```

```python
results
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 0, n_estimators = 100,
                                    criterion = 'entropy')
rf.fit(X_train, y_train)
```

```python
# Predicting Test Set
y_pred_rf = rf.predict(X_test)
acc = accuracy_score(y_test, y_pred_rf)
prec = precision_score(y_test, y_pred_rf)
rec = recall_score(y_test, y_pred_rf)
f1 = f1_score(y_test, y_pred_rf)
```

```python
df_cm = pd.DataFrame(cm, index = (0, 1), columns = (0, 1))
plt.figure(figsize = (10,7))
sns.set(font_scale=1.4)
sns.heatmap(df_cm, annot=True, fmt='g')
print("Test Data Accuracy: %0.4f" % acc)
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
model_results = pd.DataFrame([["Random Forest (n=100)", acc, prec, rec, f1]],
               columns = ["Model", "Accuracy", "Precision", "Recall", "F1 Score"])

results = results.append(model_results, ignore_index = True)
```

```python
results
```

## Feature Selection

### Recursive Feature Elimination


```python
from sklearn.feature_selection import RFE
```

```python
# Select Best X Features
rfe = RFE(rf, 20)
rfe = rfe.fit(X_train, y_train)
```

```python
# summarize the selection of the attributes
print(rfe.support_)
print(rfe.ranking_)
```

```python
X_train.columns[rfe.support_]
```

```python
# New Correlation Matrix
sns.set(style="white", font_scale=2)

# Compute the correlation matrix
corr = X_train[X_train.columns[rfe.support_]].corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(18, 15))
f.suptitle("Correlation Matrix", fontsize = 40)

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```

```python
## Randomforest
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(random_state = 0, n_estimators = 100,
                                    criterion = 'entropy')
classifier.fit(X_train[X_train.columns[rfe.support_]], y_train)

# Predicting Test Set
y_pred = classifier.predict(X_test[X_train.columns[rfe.support_]])
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

model_results = pd.DataFrame([['Random Forest after new Feature selection', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])

results = results.append(model_results, ignore_index = True)
```
